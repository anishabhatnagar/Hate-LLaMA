```python
import pandas as pd

#path to the annoation csv of HATEMM
df = pd.read_csv("HateMM_annotation.csv")
```

```python
df.columns
```

```python
prompt = "Is this hateful? Answer (Yes/No)"    
```

```python
import json
json_data = []

for index, row in df.iterrows():
    video_name = row['video_file_name']
    label = row['label']
    
    answer = 'Yes' if label == 'Hate' else 'No'
    
    video_qa = {
        "video": video_name,
        "QA": [
            {"q": prompt, "a": answer}
        ]
    }
    
    json_data.append(video_qa)

json_string = json.dumps(json_data, indent=2)
with open('instruction.json', 'w') as json_file:
    json_file.write(json_string)


print(json_string)
```
